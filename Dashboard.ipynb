```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"E:\Streamlit\Project2\Data_cleaned.csv")
```

```python
df.dtypes
```

```python
df['revenue']=df['original_price_inr']+df['delivery_charges']
```

```python
df['revenue'].fillna(0)
```

```python
#Total revenue growth rate 
Growth=df.groupby('order_year')['revenue'].sum()
Growth.plot(kind='line',marker=0,)
plt.xlabel('Year')
plt.ylabel('Revenue in INR')
plt.title('Total Revenue Growth')
plt.show()
```

```python
top_categories = df.groupby('brand')['revenue'].sum().sort_values(ascending=False).head(5)
top_categories.plot(kind='bar')
plt.title("Top Performing Categories")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```
